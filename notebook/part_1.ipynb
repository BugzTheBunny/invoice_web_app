# First we are making preperations [AKA - Part 1 - Django Rest Framework and Vue CLI

## Django
1) `pip3 install virtualenv` - so we can create a virual env
2) `virualenv {envname}` - creating virual env
3) `pip3 install django django-rest-framework django-cors-headers djoser` - Installing libraries that we will use
4) `django-admin startproject invoicely`
5) `cd invoicely`
6) **Test Django** `python manage.py runserver` -> http://localhost:8000/

## Vue
1) Install npm if you don't have it
2) `npm install -g @vue/cli` - Installing Vue
3) `vue create invoicely_vue` - creating Vue project
    - `Manually select features`
    - `Babel` + `Router` + `Vuex` + `CSS Pre-proccessors` **(remove Linter)**
    - `3.x`
    - `History mode?` - `Y`
    - `Sass/SCSS (with-dart-sass)`
    - `in dedicated config files`
    - Don't save preset
    
    
4) `cd invoicely_vue`
5) `npm install bulma` - installing Bulma (css framework)
6) `npm install axios` - installing Axios (library for talking with the backend)
7) **Test Vue** `npm run serve` -> http://localhost:8080/
